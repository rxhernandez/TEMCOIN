```python
import numpy as np
import pandas as pd
import glob

# Initialize lists to store the data
epochs = []
losses = []
accuracies = []

# Define a start flag to detect the start of the relevant data
start_processing = False
continue_processing = True  # New flag to control the end of relevant data


# Pattern to match files
file_pattern = './run10/job.res*'

# Use glob to find all files matching the pattern
file_paths = glob.glob(file_pattern)

# Iterate over all matching files
for file_path in file_paths:
    with open(file_path, 'r') as file:
        # Skip the first 1, 350, 670, 990, 1300, lines
        for _ in range(1300):
            next(file, None)
            
        while continue_processing:
            line = file.readline().strip()

            # Start processing when the first epoch line is detected
            if line.startswith('Epoch'):
                start_processing = True

            # Process the data only after the relevant data start has been detected
            if start_processing and line.startswith('Epoch'):
                parts = line.split(', Loss: ')
                epoch = int(parts[0].split()[1])
                loss = float(parts[1])
                epochs.append(epoch)
                losses.append(loss)

                # Check for the accuracy line immediately following the epoch line
                next_line = file.readline().strip()
                if next_line.startswith('Accuracy'):
                    accuracy = float(next_line.split(': ')[1].rstrip('%'))
                    accuracies.append(accuracy)
                else:
                    accuracies.append(None)
                    continue_processing = False  # Stop processing if format breaks
            else:
                # If line does not start with 'Epoch' and we are already processing, stop further processing
                if start_processing:
                    continue_processing = False

                         
# Now you have the data stored in lists:
print('Epochs:', epochs)
print('Losses:', losses)
print('Accuracies:', accuracies)

# Create a DataFrame for the results
df_results = pd.DataFrame(list(zip(epochs, losses, accuracies)), columns=['Epochs', 'Losses', 'Accuracies'])

# Write to CSV file
df_results.to_csv('output.csv', index=False)
```

```python
import re
import pandas as pd
import glob

# Base path pattern for the folders
base_folder_pattern = './run{}/job.vgg_*'

# List to store all DataFrames
all_data = []

# Process each folder
for run in range(1, 11):
    # Generate the file pattern for the current run
    file_pattern = base_folder_pattern.format(run)

    # Use glob to find all files matching the pattern
    files = glob.glob(file_pattern)

    # Process each file
    for file_path in files:
        try:
            # Read the text from the file
            with open(file_path, 'r') as file:
                # Skip the first 99 lines
                for _ in range(260):
                    next(file)
                # Read the rest of the file
                text = file.read()

            # Check if the text contains "Epoch 1, Validation Loss: "
            if "Epoch 1, Validation Loss: " in text:
                print(f"{file_path}: Epoch 1, Validation Loss: ")

                # Regex patterns to extract epochs, losses, and accuracies
                training_pattern = r'Epoch (\d+), Training Loss: ([\d.e-]+), Training Accuracy: ([\d.]+)%'
                validation_pattern = r'Epoch (\d+), Validation Loss: ([\d.e-]+), Validation Accuracy: ([\d.]+)%'

                # Extract training and validation information
                epochs_train = re.findall(training_pattern, text)
                epochs_val = re.findall(validation_pattern, text)

                # Debugging prints
                print(f"Extracted training epochs: {epochs_train}")
                print(f"Extracted validation epochs: {epochs_val}")

                # Convert to appropriate types and combine data
                for epoch in range(1, 91):  # Assuming we want to extract epochs 1 to 90
                    epoch_str = str(epoch)

                    # Find training data for the current epoch
                    train_data = next((e for e in epochs_train if e[0] == epoch_str), None)
                    val_data = next((e for e in epochs_val if e[0] == epoch_str), None)

                    data_entry = {
                        'Epoch': epoch,
                        'Training Loss': float(train_data[1]) if train_data else None,
                        'Training Accuracy (%)': float(train_data[2]) if train_data else None,
                        'Validation Loss': float(val_data[1]) if val_data else None,
                        'Validation Accuracy (%)': float(val_data[2]) if val_data else None,
                        'File Name': file_path
                    }
                    all_data.append(data_entry)

        except FileNotFoundError:
            print(f"File {file_path} not found. Please ensure the file is in the correct directory.")

# Create a DataFrame from the collected data
df = pd.DataFrame(all_data, columns=['Epoch', 'Training Loss', 'Training Accuracy (%)', 'Validation Loss', 'Validation Accuracy (%)', 'File Name'])

# Calculate the average results
average_results = df.groupby('Epoch').mean().reset_index()

# Save the concatenated DataFrame to a CSV
average_results.to_csv('average_epoch_data.csv', index=False)

# Print the DataFrame to verify
print(average_results)
```

```python
import pandas as pd
import os

# Dictionary mapping model_id (0-8) to filenames
model_filenames = {
    0: 'alexnet_trained_predictions_vmd_v4.csv',
    1: 'vgg16_trained_predictions_vmd_v4.csv',
    2: 'vgg19_trained_predictions_vmd_v1.csv',
    3: 'googlenet_trained_predictions_vmd_v5.csv',
    4: 'resnet18_trained_predictions_vmd_v2.csv',
    5: 'resnet34_trained_predictions_vmd_v2.csv',
    6: 'resnet50_trained_predictions_vmd_v2.csv',
    7: 'resnet101_trained_predictions_vmd__v2.csv',
    8: 'resnet152_trained_predictions_vmd_v2.csv'
}
models = ["AlexNet", "VGG16", "VGG19", "GoogLeNet", "ResNet18", "ResNet34", "ResNet50", "ResNet101", "ResNet152"]

# Load actual classes (replace with your actual data)
actual_classes = {
    'dimer121.png': '1QD-2origami',
    'dimer134.png': '1QD-2origami',
    'dimer144.png': '1QD-2origami',
    'dimer157.png': '1QD-2origami',
    'hexamer122.png': '1QD-6origami',
    'hexamer161.png': '1QD-6origami',
    'mono121.png': '1QD-1origami',
    'mono133.png': '1QD-1origami',
    'mono145.png': '1QD-1origami',
    'mono158.png': '1QD-1origami',
    'pentamer123.png': '1QD-5origami',
    'pentamer176.png': '1QD-5origami',
    'tetramer121.png': '1QD-4origami',
    'tetramer134.png': '1QD-4origami',
    'tetramer145.png': '1QD-4origami',
    'tetramer165.png': '1QD-4origami',
    'trimer127.png': '1QD-3origami',
    'trimer140.png': '1QD-3origami',
    'trimer158.png': '1QD-3origami',
    'trimer174.png': '1QD-3origami'
}

# Initialize a list to store the results for all models
all_model_results = []

# Loop through each model
for model_id, model_name in enumerate(models):
    filename = model_filenames.get(model_id)
    
    # Initialize an empty list to store each run's DataFrame and accuracies
    data_frames = []
    accuracies = []
    
    # Loop through each run folder
    for i in range(1, 11):
        folder_name = f'run{i}'  # Folder names are run1, run2, ..., run10
        file_path = os.path.join(folder_name, filename)  # Adjust the file name as necessary
        
        # Check if the file exists before trying to read it
        if os.path.exists(file_path):
            # Read the data file including the probability columns, ignoring initial spaces
            temp_data = pd.read_csv(file_path, header=0, skipinitialspace=True, names=['Image Name', 'Predicted Class', 'Prob1', 'Prob2', 'Prob3', 'Prob4', 'Prob5', 'Prob6'])

            # Filter only the target images
            #temp_data = temp_data[temp_data['Image Name'].isin(target_images)]

            # Drop the probability columns
            temp_data = temp_data[['Image Name', 'Predicted Class']]

            # Merge actual classes with predicted data
            temp_data['Actual Class'] = temp_data['Image Name'].map(actual_classes)

            # Create a column to check if the prediction matches the actual class
            temp_data['Correct Prediction'] = temp_data['Actual Class'] == temp_data['Predicted Class']

            # Calculate accuracy for this run
            accuracy = temp_data['Correct Prediction'].mean() * 100  # Multiply by 100 for percentage
            accuracies.append(accuracy)
            
            print(f'Prediction Accuracy for {model_name} run{i}: {accuracy:.2f}%')

            # Append the DataFrame to the list
            data_frames.append(temp_data)
        else:
            print(f"File not found: {file_path}")
    
    # Calculate the average accuracy and standard deviation across all runs for this model
    average_accuracy = sum(accuracies) / len(accuracies) if accuracies else 0
    std_accuracy = pd.Series(accuracies).std() if accuracies else 0
    model_results = {
        'Model': model_name,
        'Average Accuracy': average_accuracy,
        'Standard Deviation': std_accuracy
    }
    all_model_results.append(model_results)
    
    print(f'Average Prediction Accuracy for {model_name} across all runs: {average_accuracy:.2f}%')
    print(f'Standard Deviation of Prediction Accuracy for {model_name} across all runs: {std_accuracy:.2f}%')
    
    # Check if any DataFrames were loaded
    if data_frames:
        # Concatenate all DataFrames into one
        combined_data = pd.concat(data_frames)
    
        # Save the combined data to a CSV file
        combined_data.to_csv(f'SUM_{model_name}_predictions_vmd.csv', index=False)
    
        # Display the combined data
        print(combined_data)
    else:
        print(f"No data loaded for {model_name}. Check the files and their paths.")

# Prepare the summary DataFrame for all models
summary_data = []

for model_result in all_model_results:
    summary_data.append({
        'Model': model_result['Model'],
        'Average': model_result['Average Accuracy'],
        'Deviation': model_result['Standard Deviation']
    })

summary_df = pd.DataFrame(summary_data)

# Save the summary data to a CSV file
summary_df.to_csv('SUM_test_accuracy_vmd.csv', index=False)

# Display the summary DataFrame
print(summary_df)
```
